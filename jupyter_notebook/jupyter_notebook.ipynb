```python
# Import data
import pandas as pd

df = pd.read_csv('./winequality.csv', sep=';')
```

```python
# Rename columns
df.columns = [x.replace(" ", "_").lower() for x in df.columns]
df.head()
```

```python
# Cek Null Value
print('Total Null Value:', df.isnull().sum().sum())
```

```python
# Drop Duplicate Data
df.drop_duplicates(inplace=True)
print('DROP DUPLICATES')
print('Total Data:', df.shape[0])
print('Total Column:', df.shape[1])
print('Total Null Value:', df.isnull().sum().sum())
```

```python
# Exploratory Data Analysis Simple
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(data=df.corr().round(2), annot=True)
plt.title('Correlation of Columns')
plt.plot()
```

```python
for column in df.columns:
    sns.set_theme()
    sns.distplot(df[column])
    plt.show()
```

```python
# Scale Data
from sklearn.preprocessing import MinMaxScaler

x = df.drop(columns=['quality'])
y = df[['quality']]

scaler_x = MinMaxScaler().fit(x)
x = scaler_x.transform(x)
scaler_y = MinMaxScaler().fit(y)
y = scaler_y.transform(y)

print('Scaler X:')
print(x)
print('\nScaler Y:')
print(y)
```

```python
# Split Data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=69)
print('Size x_train:', x_train.shape)
print('Size y_train:', y_train.shape)
print('Size x_test:', x_test.shape)
print('Size y_test:', y_test.shape)
```

```python
# Make Model
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=[11]),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')
])
```

```python
# Compile Model
model.compile(
    optimizer='Adam',
    loss='mse',
    metrics=['mae']
)
```

```python
# Make Callbacks
callback_earlystopping = tf.keras.callbacks.EarlyStopping(
    min_delta=0.001,
    patience=20,
    restore_best_weights=True
)
```

```python
# Train Model
from tensorflow.keras.backend import manual_variable_initialization 

manual_variable_initialization(True)
hist = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=100,
    verbose=2,
    callbacks=[callback_earlystopping]
)
```

```python
# Plot MAE
plt.figure(figsize=(10, 6))
plt.plot(hist.history['mae'])
plt.plot(hist.history['val_mae'])
plt.xlabel('Epoch')
plt.ylabel('Mae')
plt.legend(['mae', 'val_mae'])
plt.title('Model MAE', fontsize=20)
plt.show()
```

```python
# Plot Loss
plt.figure(figsize=(10, 6))
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['loss', 'val_loss'])
plt.title('Model Loss', fontsize=20)
plt.show()
```

```python
# Test Predict
import math

data = {
    'fixed_acidity': [6.3],
    'volatile_acidity': [0.30],
    'citric_acid': [0.34],
    'residual_sugar': [1.6],
    'chlorides': [0.049],
    'free_sulfur_dioxide': [14.0],
    'total_sulfur_dioxide': [132.0],
    'density': [0.9940],
    'ph': [3.30],
    'sulphates': [0.49],
    'alcohol': [9.5]
}
df_predict = pd.DataFrame(data=data)
x_predict = scaler_x.transform(df_predict)
y_predict = model.predict(x_predict)
y_predict = math.ceil(scaler_y.inverse_transform(y_predict)[0][0])
y_predict
```
